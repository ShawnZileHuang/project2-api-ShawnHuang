## Part 2: Explore one Wikipedia page with the beautifulsoup package

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#save and print the text content of a page with all tags removed
url = 'https://en.wikipedia.org/wiki/Overseas_Chinese_Town'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
text_content = soup.get_text()
print(text_content)
```

```python
#download an image with beautifulsoup and save it in this repository
pic = requests.get('https://en.wikipedia.org/wiki/Overseas_Chinese_Town#/media/File:OCTprimaryschool.jpg')

image_name = 'pic.jpg' 
with open(image_name, 'wb') as file:
    file.write(pic.content)
pic
```

```python
#find all the links in a page with beautifulsoup
#print the first 100 characters of ten of these links
links = soup.find_all('a', href=True)

for i, link in enumerate(links[:10]):
    print(f"{link['href'][:100]}")
```

```python
#using the links in the "link" column, download the first 1000 characters of each script
#use requests and bs4, remember to remove all html tags
def get_chars(link):
    try:
        response = requests.get(link, timeout=10)
        soup = BeautifulSoup(response.content, 'html.parser')
        text = soup.get_text()
        return text[:1000]
    except Exception as e:
        return str(e)
```

```python
#add a new column to the df with the text downloaded
#save this new dataframe as "pudding_texts.csv"
scripts['Part3'] = scripts['link'].apply(get_chars)
scripts.to_csv('pudding_texts.csv', index=False)
scripts
```

## Part 4: TMDB database

#### Browse the documentation at https://developer.themoviedb.org/reference/intro/getting-started. Create an account to authenticate

```python
#create a dataset of the movies in theaters now. Include metadata fields you are interested in. 
import os
from PIL import Image
from io import BytesIO
API_KEY = '8a0076dec126bc7e193cbe34a900bc9f'
BASE_URL = 'https://api.themoviedb.org/3'

def get_now_playing_movies(api_key):
    url = f'{BASE_URL}/movie/now_playing'
    params = {
        'api_key': api_key,
        'language': 'en-US',
        'page': 1
    }
    response = requests.get(url, params=params)
    data = response.json()
    return data['results']

def download_movie_poster(movie, save_dir):
    poster_path = movie.get('poster_path')
    if poster_path:
        poster_url = f'https://image.tmdb.org/t/p/w500{poster_path}'
        response = requests.get(poster_url)
        img = Image.open(BytesIO(response.content))
        img.save(os.path.join(save_dir, f"{movie['title']}.jpg"))

save_directory = 'movie_posters'
if not os.path.exists(save_directory):
    os.makedirs(save_directory)

movies = get_now_playing_movies(API_KEY)
```

```python
#download the movie posters for 10 of these movies and save them to this repository
for movie in movies[:10]:
    download_movie_poster(movie, save_directory)
    print(f"Downloaded poster for {movie['title']}")

for movie in movies[:10]:
    print(f"Title: {movie['title']}, Release Date: {movie['release_date']}, Rating: {movie['vote_average']}")
```
